```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/skywalker290/Misleading-Video-Detection-Model/main/Datasets/sentiment-analysed.csv")
```

```python
df.columns
```

```python
df['label'] = df['LABEL'].apply(lambda x: 1 if x == 'FAKE' else 0)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score
from tqdm import tqdm

def set_threshold(numbers, threshold, labels):
    # Convert numbers to a NumPy array
    numbers = np.array(numbers)
    # Apply thresholding
    numbers = np.where(numbers >= threshold, 1, 0)
    # Compute accuracy score
    accuracy = accuracy_score(numbers, labels)
    return accuracy

# Initialize tqdm progress bar


# Dictionary to store threshold values for each parameter
thresholds = {}

# Specify the columns to compute thresholds for
columns = ['sadness', 'love', 'surprise', 'fear', 'anger', 'joy']

progress_bar = tqdm(total=len(columns), desc="Progress")

# Iterate through specified columns
for column in columns:
    array = df[column]
    max_accuracy = 0
    max_threshold = 0

    labels = df['label']

    for i in range(1,99):
        threshold = i / 100
        accuracy = set_threshold(array, threshold, labels)
        if max_accuracy < accuracy:
            max_accuracy = accuracy
            max_threshold = threshold

    # Store the maximum accuracy and corresponding threshold for the parameter
    thresholds[column] = (max_threshold, max_accuracy)

    # Update progress bar
    progress_bar.update(1)

# Close tqdm progress bar
progress_bar.close()

# Output the thresholds for specified columns
for column, (threshold, accuracy) in thresholds.items():
    print(f"Parameter: {column}, Threshold: {threshold}, Accuracy: {accuracy}")
```
