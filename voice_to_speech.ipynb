```python
import os
from pytube import YouTube
import pandas as pd

# Read the DataFrame from a CSV file
df = pd.read_csv("https://github.com/skywalker290/hack/raw/main/Datasets/Combined_dataset.csv")
# Function to download audio from YouTube video using pytube
def download_audio(video_id, output_dir):
    try:
        # Construct the YouTube video URL
        video_url = f'https://www.youtube.com/watch?v={video_id}'

        # Initialize the YouTube object with the video URL
        yt = YouTube(video_url)

        # Get the audio stream of the video
        audio_stream = yt.streams.filter(only_audio=True).first()

        output_path = os.path.join(output_dir, f'{video_id}.wav')

        # Download the audio
        audio_stream.download(output_path=output_path)
        print(f"Downloaded audio for video ID {video_id}")
    except Exception as e:
        print(f"Error downloading audio for video ID {video_id}: {e}")

# List of video IDs from the DataFrame
video_ids = df['video_id']
video_ids=video_ids[252:]
print(video_ids)

# Directory to save the downloaded audio files
output_dir = "/content/download"

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Download audio for each video ID
for video_id in video_ids:
    if video_id not in output_dir:
        download_audio(video_id, output_dir)
```

```python
import os
import speech_recognition as sr
from pydub import AudioSegment
from pydub.utils import make_chunks

def process_audio(filename,fname):
    with open(f"/content/speechtotext/{fname[:-4]}.txt", "w+") as txtf:
        myaudio = AudioSegment.from_wav(filename)
        chunks_length_ms = 8000
        chunks = make_chunks(myaudio, chunks_length_ms)
        for i, chunk in enumerate(chunks):
            chunk_name = "./chunked/" + os.path.basename(filename) + "_{0}_wav".format(i)
            print("Exporting", chunk_name)
            chunk.export(chunk_name, format="wav")
            file = chunk_name
            r = sr.Recognizer()
            with sr.AudioFile(file) as source:
                audio_listened = r.listen(source)
            try:
                rec = r.recognize_google(audio_listened)
                txtf.write(rec + ". ")
            except sr.UnknownValueError:
                print("Can't recognize audio")

            except sr.RequestError as e:
                print("Error:", e)

try:
    os.makedirs("chunked")
except FileExistsError:
    pass

directory = '/content/convertedmo4towav'

for filename in os.listdir(directory):
    if filename.endswith('.wav'):
        process_audio(os.path.join(directory, filename),filename)

```
