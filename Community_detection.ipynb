```python
import zipfile
import os
from collections import Counter

def extract_text_files(zip_file_path, extraction_path):
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(extraction_path)

def count_words_in_text_files(extraction_path):
    word_counts = Counter()
    for root, dirs, files in os.walk(extraction_path):
        for file in files:
            if file.endswith('.txt'):
                file_path = os.path.join(root, file)
                with open(file_path, 'r', encoding='utf-8') as f:
                    text = f.read()
                    words = text.split()
                    word_counts.update(words)
    return word_counts

def main(zip_file_path):
    extraction_path = 'final_comments'
    extract_text_files(zip_file_path, extraction_path)
    final_comments_path = os.path.join(extraction_path, 'final_comments')
    word_counts = count_words_in_text_files(final_comments_path)

    # Find the first 20 most recurring words
    most_common_words = word_counts.most_common(20)
    print("The first 20 most recurring words are:")
    for word, count in most_common_words:
        print(f"'{word}': {count} occurrences.")

zip_file_path = '/content/final_comments.zip'
main(zip_file_path)
```

```python
import zipfile
import os
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt


def count_words_in_text_files(extraction_path):
    word_counts = Counter()
    for root, dirs, files in os.walk(extraction_path):
        for file in files:
            if file.endswith('.txt'):
                file_path = os.path.join(root, file)
                with open(file_path, 'r', encoding='utf-8') as f:
                    text = f.read()
                    words = text.split()
                    word_counts.update(words)
    return word_counts

def main(zip_file_path, label):
    extraction_path = 'final_comments'
    extract_text_files(zip_file_path, extraction_path)
    final_comments_path = os.path.join(extraction_path, 'final_comments')
    word_counts = count_words_in_text_files(final_comments_path)

    # Filter words based on the label
    label_words = [word for word in word_counts.keys() if label in word.lower()]
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(label_words))

    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'Word Cloud for Label "{label}"')
    plt.show()

zip_file_path = '/content/final_comments.zip'
main(zip_file_path, label='FAKE')
```

```python
import os
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter

def read_text(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    words = text.split()
    return words

directory = '/content/final_comments/final_comments/'
all_words = []
for filename in os.listdir(directory):
    if filename.endswith('.txt'):
        file_path = os.path.join(directory, filename)
        words = read_text(file_path)

        all_words.extend(words)

word_counts = Counter(all_words)
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_counts)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import networkx as nx

data = pd.read_csv("/content/Combined_dataset.csv")

tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(data['title'])
cosine_sim_titles = cosine_similarity(tfidf_matrix, tfidf_matrix)

G_titles = nx.Graph()
for i in range(len(data)):
    for j in range(i+1, len(data)):
        if cosine_sim_titles[i][j] > 0.5:  # Threshold for similarity
            G_titles.add_edge(i, j)

# Detect communities using Louvain method
communities_titles = nx.algorithms.community.greedy_modularity_communities(G_titles)

for i, community in enumerate(communities_titles):
    print(f"Community {i+1}: {community}")
    # Print the nodes and edges of the graph G_titles

import matplotlib.pyplot as plt
pos = nx.kamada_kawai_layout(G_titles, weight='weight')  # Position nodes using Kamada-Kawai layout with considering edge weights
nx.draw(G_titles, pos, with_labels=False, node_color='skyblue', node_size=100, edge_color='black', linewidths=1, font_size=10)

plt.title("Graph of Video Titles")
plt.show()


fake_counts = {}
real_counts = {}

for i, community in enumerate(communities_titles):
    fake_count = sum(1 for node in community if data.loc[node, 'LABEL'] == 'FAKE')
    real_count = sum(1 for node in community if data.loc[node, 'LABEL'] == 'REAL')
    fake_counts[i+1] = fake_count
    real_counts[i+1] = real_count

#ratio of fake and real videos in each community
ratios = {}
for i, community in enumerate(communities_titles):
    ratio = fake_counts[i+1] / (fake_counts[i+1] + real_counts[i+1])
    ratios[i+1] = ratio
for community_id, ratio in ratios.items():
    print(f"Community {community_id}: Ratio of fake videos = {ratio:.2f}, Ratio of real videos = {1 - ratio:.2f}")
```

```python
# Compute TF-IDF vectors for video descriptions
tfidf_matrix_desc = tfidf_vectorizer.fit_transform(data['description'].fillna(''))

# Compute cosine similarity between description vectors
cosine_sim_desc = cosine_similarity(tfidf_matrix_desc, tfidf_matrix_desc)

# Build a graph using the cosine similarity matrix
G_desc = nx.Graph()
for i in range(len(data)):
    for j in range(i+1, len(data)):
        if cosine_sim_desc[i][j] > 0.5:  # Threshold for similarity
            G_desc.add_edge(i, j)

# Detect communities using Louvain method
communities_desc = nx.algorithms.community.greedy_modularity_communities(G_desc)

# Print communities
for i, community in enumerate(communities_desc):
    print(f"Community {i+1}: {community}")
import matplotlib.pyplot as plt

# Draw the graph using NetworkX and Matplotlib
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G_desc)
nx.draw(G_desc, pos, with_labels=False, node_size=50)

# Calculate and print the ratio of real and fake videos in each community
ratios = {}
for i, community in enumerate(communities_desc):
    fake_count = sum(1 for node in community if data.loc[node, 'LABEL'] == 'FAKE')
    real_count = sum(1 for node in community if data.loc[node, 'LABEL'] == 'REAL')
    ratio = fake_count / (fake_count + real_count)
    ratios[i+1] = ratio
    print(f"Community {i+1}: Ratio of fake videos = {ratio:.2f}, Ratio of real videos = {1 - ratio:.2f}")

plt.title("Graph of Video Descriptions with Community Detection")
plt.show()
```

```python
# Compute the difference of views/likes ratio
weights = {}
for edge in G_desc.edges():
    node1, node2 = edge
    views_likes_ratio1 = data.loc[node1, 'views'] / data.loc[node1, 'likes']
    views_likes_ratio2 = data.loc[node2, 'views'] / data.loc[node2, 'likes']
    weight = abs(views_likes_ratio1 - views_likes_ratio2)
    weights[edge] = weight

# Add weights to the graph
nx.set_edge_attributes(G_desc, weights, 'weight')

# Community detection using Louvain method
communities_louvain = nx.algorithms.community.greedy_modularity_communities(G_desc, weight='weight')
# Print communities and videos for Louvain method
print("Louvain Method:")
for i, community in enumerate(communities_louvain):
    video_ids = [node for node in community]
    print(f"Community {i+1}: Video IDs = {video_ids}")

# Community detection using Newman-Girvan method
communities_newman_girvan = nx.algorithms.community.girvan_newman(G_desc)

# ratio of real and fake videos in each community for Louvain method
print("Louvain Method:")
for i, community in enumerate(communities_louvain):
    fake_count = sum(1 for node in community if data.loc[node, 'LABEL'] == 'FAKE')
    real_count = sum(1 for node in community if data.loc[node, 'LABEL'] == 'REAL')
    ratio = fake_count / (fake_count + real_count)
    print(f"Community {i+1}: Ratio of fake videos = {ratio:.2f}, Ratio of real videos = {1 - ratio:.2f}")

# Community detection using Newman-Girvan method
communities_newman_girvan = next(nx.algorithms.community.girvan_newman(G_desc))
print("\nNewman-Girvan Method:")
for i, community in enumerate(communities_newman_girvan):
    print(f"Community {i+1}: {community}")
# Calculate ratio for Newman-Girvan method
print("\nNewman-Girvan Method:")
for i, community in enumerate(communities_newman_girvan):
    fake_count = sum(1 for node in community if data.loc[node, 'LABEL'] == 'FAKE')
    real_count = sum(1 for node in community if data.loc[node, 'LABEL'] == 'REAL')
    ratio = fake_count / (fake_count + real_count)
    print(f"Community {i+1}: Ratio of fake videos = {ratio:.2f}, Ratio of real videos = {1 - ratio:.2f}")
import matplotlib.pyplot as plt

# # Draw the graph
# plt.figure(figsize=(10, 6))
# pos = nx.spring_layout(G_desc)  # Layout algorithm
# nx.draw(G_desc, pos, with_labels=True, node_color='skyblue', node_size=1000, edge_color='black', linewidths=1, font_size=10)
# labels = nx.get_edge_attributes(G_desc, 'weight')
# nx.draw_networkx_edge_labels(G_desc, pos, edge_labels=labels)

# # Show the plot
# plt.title("Graph Visualization")
# plt.show()
```
